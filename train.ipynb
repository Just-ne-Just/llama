```python
from src.dataset.dataset import TinyStoriesDataset
from src.model.llama import LLAMA
from src.model.utils import create_mask
from src.train.train import train
```

```python
train_ds = TinyStoriesDataset(data_file="/kaggle/input/tiny-stories-ds/TinyStoriesV3-GPT4-train.txt", sp_model_prefix="/kaggle/working/llama/src/tokenizer/bpe/bpe")
```

```python
val_ds = TinyStoriesDataset(data_file="/kaggle/input/tiny-stories-ds/TinyStoriesV3-GPT4-valid.txt", sp_model_prefix="/kaggle/working/llama/src/tokenizer/bpe/bpe")
```

```python
from torch.utils.data import DataLoader
```

```python
len(train_ds)
```

```python
train_loader = DataLoader(train_ds, batch_size=64, shuffle=True)
```

```python
val_loader = DataLoader(val_ds, batch_size=64, shuffle=False)
```

```python
import gc
import torch

gc.collect()
torch.cuda.empty_cache()
```

```python
model = LLAMA(num_layers=2, embed_dim=32, num_heads=2, vocab_size=train_ds.vocab_size, ff_dim=32, dropout=0.1)
```

```python
model = model.to('cuda')
```

```python
from torch.optim import AdamW
```

```python
optimizer = AdamW(model.parameters(), lr=3e-4, weight_decay=1e-5)
```

```python
device = 'cuda'
```

```python
train(5, model, train_ds.pad_id, optimizer, train_loader, val_loader, device, 5000)
```
