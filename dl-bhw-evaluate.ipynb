```python
from src.dataset.dataset import TinyStoriesDataset
from src.model.llama import LLAMA
import torch
from src.train.train import generate
```

```python
model = LLAMA(num_layers=8, embed_dim=512, num_heads=8, vocab_size=ds.vocab_size, ff_dim=2048, dropout=0.1)
```

```python
pretrained_path = "/kaggle/input/bhw-checkpoint/checkpoint_best.pth"
device = 'cuda:0'
```

```python
checkpoint = torch.load(pretrained_path, device)
model.load_state_dict(checkpoint["state_dict"])
```
